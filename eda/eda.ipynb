```python
import pandas as pd



# Step 1: Detect Missing Values
missing_values = df.isnull().sum()
print("Missing Values:\n", missing_values)

# Step 2: Detect Outliers using the IQR method
def detect_outliers_iqr(df):
    outlier_indices = []
    for column in df.select_dtypes(include=['float64', 'int64']).columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Identify outliers
        outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
        outlier_indices.extend(outliers.index.tolist())
        
        print(f"\nOutliers detected in '{column}':")
        print(outliers)

    # Remove duplicate indices (if an index is an outlier for multiple columns)
    return set(outlier_indices)

outlier_indices = detect_outliers_iqr(df)

# Print the indices of detected outliers
print("\nIndices of outliers:", outlier_indices)

```

```python
# Correlation Matrix (linear relationships)
print("\nCorrelation matrix of features and target:")
corr_matrix = df.corr()

# Visualize correlations via a heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, cmap='coolwarm', center=0, linewidths=1)
plt.title('Correlation Heatmap')
plt.show()

# Finding features with high correlation (threshold can be adjusted)
threshold = 0.6
high_corr_pairs = [(i, j) for i in corr_matrix.columns for j in corr_matrix.columns 
                   if abs(corr_matrix[i][j]) > threshold and i != j]
print("Highly correlated pairs:")
for pair in high_corr_pairs:
    print(pair)
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from scipy.stats import spearmanr
from sklearn.feature_selection import mutual_info_regression
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Normalize (Standardize) all features to the same range
scaler = StandardScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)

# Step 2: Calculate Spearman's rank correlation matrix
spearman_corr_matrix = df_scaled.corr(method='spearman')

# Step 3: Calculate Mutual Information for each feature against the target
X = df_scaled.drop(columns=['target']) 
y = df_scaled['target']
mutual_info = mutual_info_regression(X, y)

# Display top Spearman correlations with the target
print("Top Spearman's Rank Correlations with the Target:")
print(spearman_corr_matrix['target'].sort_values(ascending=False))

# Display Mutual Information Scores for all features
mutual_info_series = pd.Series(mutual_info, index=X.columns).sort_values(ascending=False)
print("\nMutual Information Scores:")
print(mutual_info_series)

# Step 4: Visualize Spearman's correlation matrix using a heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(spearman_corr_matrix, cmap='coolwarm', annot=False, center=0, linewidths=1)
plt.title('Spearman Correlation Heatmap')
plt.show()

# Optional: Display features with high mutual information (above a certain threshold)
threshold = 0.1  
high_mutual_info_features = mutual_info_series[mutual_info_series > threshold]
print("\nFeatures with High Mutual Information (Threshold > 0.1):")
print(high_mutual_info_features)
```

Spearman's correlation measures monotonic relationships (whether linear or non-linear) between ordinal or continuous variables, but still shows weak correlations between most features and the target, except for a few (like Feature 7), but even thought a little.
Pearson correlation has no sense to try - no normal distribution.


Mutual information is a distance between two probability distributions. Mutual information helps capture both linear and non-linear relationships. The top features with the highest mutual information scores with the target are:

Feature 6: 4.6021 (this feature stands out with a much higher score than others, indicating a stronger relationship with the target)

Feature 7: 1.2742 (this also shows a relatively stronger relationship with the target)

All other features have near-zero mutual information scores, indicating weak or no relationships.
